# Decentralized Identifiers (DIDs) v1.0

## Caso de USO:


1. Alice tem 3 DIDs, um para fazer compras no ebay, outro para usar em redes sociais e um terceiro para interagir com o banco.

dividir em partes: <br/>
* Alice tem 3 DIDs
    * dados: forma de armazenamento e segurana


* compras no ebay
    * dados: comprovao de registro no site e da forma de pagamento

* redes sociais
    * dados: nome, idade, email, gerar e armazenar certificado de acesso

* interagir com o banco
    * dados: comprovao de existncia, gerar e armazenar certificado de acesso 

Para o caso a que seria Alice ter 3 DIDs(que podem ser DIDs subject, DIDs document e uma DID URL) primeiramente ela teria que ter um meio para armazenar tais DIDs, uma DID controller, que poderiam ser por meio de uma carteira virtual assim como as wallets de cripto moedas e, assim, utilizar suas DIDs.


para o caso b teria que ter um certificado gerado por um banco confivel, assim, podendo gerar uma DID focada em dados monetrios que confiana tanto para o comprador quanto para o vendedor.

para o caso c a DID de acesso as redes poderiam ser mais simples e geradas por um rgo federal, pois, por conter dados como idade, nome e email, poderia ser contedo previamente registrado em tal rgo, e aps o cadastro em tal rede, poderia ser gerado uma DID da rede social, tal como um QR code para o acesso.


para o caso d aps ter o cadastro no banco usando sua DID que comprova sua forma de pessoa, fsica ou jurdica, poderia se gerar uma DID prpria do banco para controlar o acesso e o uso.

```python
from random import randint
# vou colocar um modelo em json

DIDsBank = {}

def generation(name):
    name = name
    did = randint(10000000,99999999)
    DIDsBank[name] = did
    print(f"{name}, sua DID gerada foi {did}.")
    return did

def verification(did):
    for i in DIDsBank.values():
        if did == i:
            print("DID verificada e autenticada.")
        else:
            print("DID no encontrada.")

nome = input("Nome:")
            
myDID = generation(nome)

print(myDID)

verification(myDID)
```
